```python
import numpy as np
import pandas as pd
import math
import seaborn as sb
import matplotlib.pyplot as plt
from scipy import stats
```

```python
track_data = pd.read_csv("../data/top-200-dataset/spotify-top-200-dataset.csv", index_col=0, delimiter=';', parse_dates=['week'])
track_data.columns
tracks = track_data[['track_name', 'track_popularity', 'album_popularity', 'artist_names', 'artist_popularity', 'rank', 'explicit', 'danceability', 'energy', 'key', 'time_signature',
'loudness', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'duration', 'week'
]]
tracks = tracks.reset_index()
```

```python
top = tracks.reset_index().sort_values('track_popularity', ascending=False).drop_duplicates('track_id').set_index('track_id')
```

```python
features = ['explicit', 'danceability',
       'energy', 'key', 'time_signature', 'loudness', 'speechiness',
       'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo',
       'duration']


# top = top[top['track_popularity'] >= 80]
rs = []
for feature in features:
       r = stats.linregress(top['track_popularity'], top[feature]).rvalue
       rs.append((feature, r))
    

rData = pd.DataFrame(rs, columns=['feature', 'correlation'])
rData.sort_values('correlation', ascending=False)
```
